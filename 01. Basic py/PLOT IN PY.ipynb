## Plot dengan seaborn


**Templet Plot **

```
sns.jenis_plot(data=nama_dataframe, x='variabel_x', y='variabel_y', hue='variabel_kategori', style='variabel_gaya', color='warna', palette='skema_warna')
```

**Penjelasan** 

`hue` : Ini memisahkan warna, gunakan variable untuk memisahkan warna (contoh dipisahkan berdasarkan warna jenis kelamin)

`style` : Variabel kategori yang akan memisahkan data berdasarkan warna. (Opsional)

`palette` : Skema warna untuk variabel `hue`. ('Set1','Set2','coolwarm','viridis','Blues')

---
___

|Jenis Plot|Gunakan Untuk|x/y|hue/style||
|-|-|-|-|-|

## Plot dengan plt

### Line plot

`plt.plot(x, y, marker, linestyle, color)`

Line Plot (Grafik garis) adalah jenis grafik yang digunakan untuk menunjukan perubahan suatu nilai dari waktu kewaktu atau tren berdasarkan urutan tertentu.

|Jenis|isinya|Kegunaan|
|-|-|-|
|marker||digunakan untuk menandai setiap titik data pada grafik|
||o (huruf o)|menggunakan simbol lingkaran pada setiap titik|
||.|simbol titik titik pada setiap titik|
||x|simbol silang pada setiap titik|
||^|simbol segitiga pada setiap titik|
|linestyle||Jenis garis pada plot|
||-|garis lurus|
||:|garis titik titik|
||--|garis putus putus|
||-.|Kombinasi garis titik dan garis|
|color||Warna garis|
||b|blue|
||r|red|
||g|green|
||k|black|

````python
plt.figure(figsize=(10,4)) # figure object yang akan menjadi kanvas utama tempat grafik digambar.
plt.plot(Data_plot['State'], Data_plot['Total day minutes'], marker = '^', linestyle = ':', color ='blue')

plt.xlabel('State')
plt.ylabel('Tital Dat minutes')
plt.title('Line Plot')

plt.savefig('plot_line.png', dpi =300, bbox_inches = 'tight')

plt.show()
````

### Bar Plot

plt.bar(kategori, nilai, color='skyblue')

````python
# buat data Churn di plot barnya

Data_churn = Data_plot['Churn'].value_counts()
Data_churn.head(5)
````

````python
# Pastikan indexnya adladh false dulu atau true dulu
Data_churn.reindex([True, False])
````

````python
# Membuat plot
plt.figure(figsize=(10,4))
plt.bar(['Tidak Churn', 'Churn'], Data_churn.values, color =['red', 'blue'])

plt.title('Bar Plot untuk Churn', fontsize=20, loc='center')
plt.xlabel('Status', fontsize=15, rotation=0)
plt.ylabel('Total', fontsize=15, rotation=90)
plt.xticks(rotation =45, fontsize=15)
plt.yticks(fontsize=15)

# Menambahkan nilai
for i, value in enumerate(Data_churn):
  plt.text(i, value+0.5, str(value), ha='center', va ='bottom', fontsize=10)

# Menyiimpn file dalam bentuk png
plt.savefig('churn_barplot.png',dpi =300, bbox_inches = 'tight')

plt.show()
````

### Diagram flow




````python
from graphviz import Digraph
from IPython.display import display, Image, SVG, display_pdf, IFrame

# Membuat objek diagram
dot = Digraph(comment='Profit Decision Tree')

# Node awal
dot.node('A', 'All Data', shape ='box')

# Cabang pertama
dot.node('B', 'Profit <= 0', shape ='box')
dot.node('C', 'Profit > 0', shape ='box')

# Cabang kedua
dot.node('D', 'data_3', shape ='box')
dot.node('E', 'Profit >= Cost', shape ='box')
dot.node('F', 'Profit < Cost', shape ='box')

# Hasil akhir
dot.node('G', 'data_1', shape ='box')
dot.node('H', 'data_2', shape ='box')

# Hubungkan node
dot.edges(['AB', 'AC', 'BD', 'CE', 'CF', 'EG', 'FH'])

# Simpan diagram ke file
output_path = dot.render('profit_tree', format='pdf', cleanup=True)

#Tampilkan di colab
display(IFrame(output_path, width=600, height=400))
````
