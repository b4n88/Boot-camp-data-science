# Pandas

Apa yang bisa dilakukan dengan Pandas?

```python
# Membuat series
s = pd.Series([1,2,3,4,5,6])

# Membuat  Data frame (Gabungan dari beberpa series)
df =pd.DataFrame({
    'Nama': ['Andi', 'Budi', 'Citra'],
    'Usia': [23, 25, 22],
    'Kota': ['Jakarta', 'Bandung', 'Surabaya']
})

# Membaca file
df = pd.read_csv('nama file.csv') # Bisa dengan excel sql jason juga

# Mengeksppr data
df.to_csv('output.csv', index = False)

# Eksplorasi
print(df.head()) # Data atas
print(df.tail()) # Data bawah

# Informsi data
df.info()

# Deskriptif data
df.describe()

# Melihat kolom
df.columns

# Dimensi data
df.shape

# seleksi kolom
df['kd toko'] # seleksi 1 kolom
df[['nama', 'Usia']] # beberapa kolom

#hapus kolom
df = df.drop('Kota', axis=1)  # Hapus kolom

#hapus baris
df = df.drop(0, axis = 0)

#hapus nilai kosong
df.dropna()

# MEngisi nilai kosong
df.fillna(0)
df['Harga'].fillna(df['Harga'].meam(), inplace = True)

# Duplikat
df.duplicated().sum()
df.duplicated(subset = ['Nama']).sum()

# mengelompokan
df_group = df.groupby('kota').mean()

#pivot table
df_pivot = df.pivot_table(index = 'kota', value = 'gaji', aggfunc = 'mean')

#Short data
df.sprt_values(by = usia, ascending = True)

# Pengelolaan string
df['Nama'] = df['Nama'].str.upper() # huruf kapitalsmua
df['Nama'] = df['Nama'].str.title() # Huruf awal menjadi besar semua
df['Nama'] = df['Nama'].str.capitalize() # Huruf pertama kapital
df['Nama'] = df['Nama'].str.strip() # menghaous sppasi di awal dan di akhir str
df['Nama'] = df['Nama'].str.lstrip() # Menghapus awal spasi 
df['Nama'] = df['Nama'].str.rstrip() # Menghapus akhir spasi
df['Nama'] = df['Nama'].str.replace('Andi', 'Yudi') # Mengganti substring
df['Nama'] = df['Nama'].str.contains('yuda') # Melihat apakah ada yuda?
df['Nama'] = df['Nama'].str.len() # menghitung panjang
df['Nama'] = df['Nama'].str.isdigit() # memeriksa apakah berisi angka?
df['Nama'] = df['Nama'].str.isalpha() # Memeriksa apakah berisi huruf?

df['status']= df['status'].replace({True = 'Ya', False = 'Tidak'})

# Statistika
df['Usia'].mean()
df['Usia'].median()
df['Usia'].std()
df['Usia'].sum()
df['Usia'].count()

# korelasi antar kolom
df.corr()

# format waktu
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
df['Bulan'] = df['Tanggal'].dt.month
df['Bulan'] = df['Tanggal'].dt.month_month()
df['Nama_Bulan'] = df['Tanggal'].dt.strftime('%B') #Bulan lengkap
df['Nama_Bulan'] = df['Tanggal'].dt.strftime('%b') #Bulan 3 digit

#Menambah data
df['Bulan'] = df['Tanggal'].dt.month
df['Nama_Bulan'] = df['Tanggal'].dt.month_name()

```
